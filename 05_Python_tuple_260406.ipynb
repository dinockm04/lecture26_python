### 튜플 만들기

```python
# 튜플 생성 함수

# 1. 빈 튜플 생성
def create_empty_tuple():
    return (), tuple()

# 2. 요소 1개 튜플 생성
def create_single_tuple(value):
    return (value,)

# 3. 여러 요소 튜플 생성 (packing)
def create_tuple(*args):
    return args

# 4. 튜플 언패킹 함수
def unpack_tuple(t):
    return t


# 실행 예시
t4, t5 = create_empty_tuple()
print("빈 튜플:", t4, t5)

t_single = create_single_tuple(1)
print("요소 1개 튜플:", t_single)

t6 = create_tuple(1, 2, 3)
print("튜플 생성:", t6)

# 언패킹
x, y, z = unpack_tuple(t6)
print("언패킹 결과:", x, y, z)
```

### 튜플 다루기

```python
# 튜플 다루기 함수

# 1. 튜플 항목 접근
def access_tuple(t):
    print("인덱싱:", t[0])
    print("슬라이싱:", t[1:3])
    
    print("for문:")
    for item in t:
        print(item, end=' ')
    print()


# 2. 튜플 탐색
def search_tuple(t, value):
    print("in 연산:", value in t)
    if value in t:
        print("index:", t.index(value))
        print("count:", t.count(value))
    else:
        print("값 없음")


# 3. 튜플 정보
def tuple_info(t):
    print("길이:", len(t))
    print("최댓값:", max(t))
    print("최솟값:", min(t))
    print("합계:", sum(t))


# 4. 새로운 튜플 생성
def create_new_tuple(t):
    print("정렬:", tuple(sorted(t)))
    print("덧셈:", t + (100, 200))
    print("곱셈:", t * 2)


# 5. 튜플 삭제
def delete_tuple():
    t = (1, 2, 3)
    del t
    print("튜플 삭제 완료")


# 실행 예시
t = (3, 1, 2, 1)

access_tuple(t)
print()

search_tuple(t, 1)
print()

tuple_info(t)
print()

create_new_tuple(t)
print()

delete_tuple()
```

### 튜플의 활용

```python
# 튜플 활용 함수

# 1. swap (값 교환)
def swap_values(x, y):
    x, y = y, x
    return x, y


# 2. 여러 값 반환 (튜플 자동 생성)
def add_and_mul(a, b):
    return a + b, a * b


# 3. 데이터 무결성 예시 (변경 불가)
def immutable_example():
    point = (37.5, 127.0)  # GPS 좌표
    print("좌표:", point)
    # point[0] = 0  ❌ 오류 발생 (수정 불가)


# 4. 딕셔너리 키로 사용
def tuple_as_key():
    d = {}
    key = (1, 2)
    d[key] = "튜플 키 가능"
    print(d)


# 5. 속도/읽기용 데이터
def read_only_data():
    data = (1, 2, 3, 4, 5)
    for i in data:
        print(i, end=' ')
    print()


# 실행 예시
print("swap:", swap_values(10, 20))

result = add_and_mul(3, 5)
print("합과 곱:", result)

immutable_example()

tuple_as_key()

read_only_data()
```
