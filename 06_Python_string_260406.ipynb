### 문자열 분리와 결합하기

```python
# 1. split() - 문자열 분리
def split_string(s):
    result = s.split(',')
    print("분리 결과:", result)
    return result


# 2. join() - 문자열 결합
def join_string(lst):
    result = ','.join(lst)
    print("결합 결과:", result)
    return result


# 실행 예시
s = "KAIST,제주대,홍익대"
lst = split_string(s)

join_string(lst)
```

### 공백 제거와 문자열 변경하기

```python
# 1. strip() - 양쪽 공백/문자 제거
def strip_example(s):
    print("strip:", s.strip())


# 2. rstrip() - 오른쪽 제거
def rstrip_example(s):
    print("rstrip:", s.rstrip('kg'))


# 3. lstrip() - 왼쪽 제거
def lstrip_example(s):
    print("lstrip:", s.lstrip('#'))


# 4. replace() - 문자열 변경
def replace_example(s):
    print("replace:", s.replace('대', '대학교'))


# 실행 예시
s1 = "  Hi! Python  "
s2 = "..Hi! Python.."
s3 = "1234kg"
s4 = "###student_name"
s5 = "KAIST,제주대,홍익대"

strip_example(s1)
strip_example(s2)

rstrip_example(s3)

lstrip_example(s4)

replace_example(s5)
```

### 문자열 찾기

```python
# 1. find() - 왼쪽부터 찾기
def find_example(s, word):
    result = s.find(word)
    print("find 결과:", result)
    return result


# 2. rfind() - 오른쪽부터 찾기
def rfind_example(s, word):
    result = s.rfind(word)
    print("rfind 결과:", result)
    return result


# 3. count() - 개수 세기
def count_example(s, word):
    result = s.count(word)
    print("count 결과:", result)
    return result


# 실행 예시
s = "Hi! Python is Python"

find_example(s, "Python")
rfind_example(s, "Python")
count_example(s, "Python")
```

### 형 변환하기

```python
# 1. str() - 숫자 → 문자열
def to_string(num):
    result = str(num) + "kg"
    print("str 결과:", result)
    return result


# 2. int() - 문자열 → 정수
def to_int(s):
    num = int(s)
    print("짝수 여부:", num % 2 == 0)
    return num


# 3. float() - 문자열 → 실수
def to_float(s):
    result = float(s) / 2
    print("float 결과:", result)
    return result


# 4. cm 처리 (시험 핵심 코드)
def convert_height(s):
    if s.find("cm") != -1:
        height = float(s.rstrip("cm"))
    else:
        height = float(s)
    print("키:", height)
    return height


# 실행 예시
to_string(1234)

to_int("6")

to_float("12.34")

convert_height("180cm")
convert_height("175")
```

### 문자열 구성 확인하기

```python
def check_string(s):
    print("isdigit:", s.isdigit())
    print("isalpha:", s.isalpha())
    print("isalnum:", s.isalnum())
    print("islower:", s.islower())
    print("isupper:", s.isupper())
    print("isspace:", s.isspace())


# 실행 예시
check_string("1234")
check_string("python")
check_string("pass123")
check_string("PYTHON")
check_string(" ")
```

### 대소문자 확인하기

```python
def convert_case(s):
    print("lower:", s.lower())
    print("upper:", s.upper())


# 실행 예시
convert_case("Python")
```
