### 리스트(list) 만들기

```python
a = list()
b = list([1, 2, 3])
c = list('hello')
```

### 리스트 항목에 접근하기

```python
test_list = [1, 2, 3, 'one', 'two', 'three']

print(test_list[0])
print(test_list[3])
print(test_list[-1])
print(test_list[-3])

for item in test_list:
    print(item)
```

### 리스트의 항목에 접근하기 : 슬라이싱(slicing)

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]

print(test_list[0:5])
print(test_list[5])

print(test_list[1:3])
print(test_list[:5:2])
print(test_list[::3])
print(test_list[:])
print(test_list[::])

print(test_list[-1:-7:-2])
print(test_list[1:-1:])
print(test_list[-3::-1])
print(test_list[])
```

### 리스트에 항목 추가하기

```python
a_list = [1, 2, 3, 4]

a_list.append('four')
print(a_list)

a_list.insert(3, 'three')
print(a_list)

a_list.extend([10, 11])
print(a_list)

b_list = [1, 2, 3]
c_list = ['a', 'b', 'c']
d_list = b_list + c_list
print(d_list)

e_list = b_list * 3
print(e_list)
```

### 리스트에서 항목 삭제하기

```python
test_list = [1, 2, 3, 'a', 'b', 'c', 10, 11]

item = test_list.pop(2)
print(item)
print(test_list)

item = test_list.pop()
print(item)
print(test_list)

test_list.remove('a')
print(test_list)

del test_list[2]
print(test_list)

del test_list[:2]
print(test_list)

test_list.clear()
print(test_list)
```

### 리스트 항목 재배열하기

```python
a_list = ['one', 'two', 'three', 'four']

a_list.reverse()
print(a_list)

a_list.sort()
print(a_list)

b_list = sorted(a_list, reverse=True)
print(b_list)

a_list.sort(key=len)
print(a_list)
```

### 리스트 탐색하기

```python
test_list = [1, 2, 3, 1, 2, 3]

target = 3
print(target in test_list)
print(test_list.index(target))
print(test_list.count(target))

target = 4
print(target in test_list)
print(test_list.count(target))

print(len(test_list))
print(max(test_list))
print(min(test_list))
print(sum(test_list))
```

### 리스트 내포, list comprehension

```python
sequence = [x for x in range(1, 101)]
print(sequence)

sequence = [x for x in range(1, 101) if x % 7 == 0]
print(sequence)
```
